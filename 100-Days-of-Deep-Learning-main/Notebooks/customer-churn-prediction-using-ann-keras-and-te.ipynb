```python
df = pd.read_csv('/kaggle/input/credit-card-customer-churn-prediction/Churn_Modelling.csv')
```

```python
df.head()
```

```python
df.drop(columns = ['RowNumber','CustomerId','Surname'],inplace=True)
```

```python
df.head()
```

```python
df['Geography'].value_counts()
```

```python
df['Gender'].value_counts()
```

### One-hot Encode the Categorical Columns

```python
df = pd.get_dummies(df,columns=['Geography','Gender'],drop_first=True)
```

```python
df.head()
```

```python
X = df.drop(columns=['Exited'])
y = df['Exited'].values

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential 
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()

model.add(Dense(3,activation='sigmoid',input_dim=11))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

[MLP](https://imgur.com/a/XDeHrCW)

```python
model.compile(optimizer='Adam',loss='binary_crossentropy')
```

```python
model.fit(X_train_scaled,y_train, epochs = 10)
```

### Improving Accuracy
### 1.Increasing Hidden Layer
### 2.Changing *activation* = **Relu**
### 3.Increasing *Epoch Values*
### 4.Increasing *Node* in Hidden Layer

```python
model = Sequential()

model.add(Dense(11,activation='relu',input_dim=11))
model.add(Dense(11,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

### Our target is to observer both accuracy & val_accuracy are increasing or not
### And both the loss & Validation_loss is reducing or not

> #### Here slight overfitting, as training_loss reduced slight more than validation_loss

```python
model.compile(optimizer='Adam',loss='binary_crossentropy' , metrics = ['accuracy']) #It shows Accuracy Alongside
```

```python
model.fit(X_train_scaled,y_train, epochs = 100 , validation_split = 0.2) #Validation split means seperating 20%
```

```python
model.layers[0].weights
```

```python
model.layers[1].weights
```

```python
model.layers[2].weights
```

```python
y_log = model.predict(X_test_scaled)
```

```python
y_pred = np.where(y_log > 0.5 , 1 , 0)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
