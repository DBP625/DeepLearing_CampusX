```python
df.drop(columns = ['Serial No.'] , inplace = True)
```

```python
X = df.iloc[: , 0:-1]
y = df.iloc[: , -1]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test , y_train , y_test = train_test_split(X , y , test_size = 0.2 , random_state = 1)
```

```python
X_train
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled  =  scaler.fit_transform(X_test)
```

```python
import tensorflow
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense
```

Using `activation='linear'` means:
- The model outputs **a real number**, without squashing it through any function (like `sigmoid` or `softmax`).
- It's typically used for **regression problems**.

### ðŸ§  ANN Structure Breakdown

**Input Layer**  
- Takes 7 features (input_dim = 7)

**Hidden Layer**  
- 7 neurons  
- Activation: ReLU  
- Parameters = 7Ã—7 (weights) + 7 (biases) = **56**

**Output Layer**  
- 1 neuron  
- Activation: Linear (no transformation, ideal for regression)  
- Parameters = 7Ã—1 (weights) + 1 (bias) = **8**

---

```python
model = Sequential()
model.add(Dense(7 , activation = 'relu' , input_dim = 7))
model.add(Dense(7 , activation = 'relu'))
model.add(Dense(1 , activation = 'linear'))
```

```python
model.summary()
```

```python
model.compile(loss = 'mean_squared_error' , optimizer = 'Adam')
```

```python
history =  model.fit(X_train_scaled , y_train , epochs = 100 , validation_split = 0.2) 
```

```python
y_pred = model.predict(X_test_scaled)
```

### Very bad r2_score here
### before adding another hidden layer
### before increasing the epochs

```python
from sklearn.metrics import r2_score
r2_score(y_test , y_pred)
```

### Almost the val_loss matches the actual loss

```python
import matplotlib.pyplot as plt
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
```
